```python
import pandas as pd
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
miles_decomp_df = pd.read_csv('us-airlines-monthly-aircraft-miles-flown.csv', parse_dates=[0])
```

```python
miles_decomp_df.index = miles_decomp_df['Month']
```

```python
result = seasonal_decompose(miles_decomp_df['MilesMM'], model = 'additive')
```

```python
result.plot()
```

```python
result2 = seasonal_decompose(miles_decomp_df['MilesMM'], model = 'multiolicative')
```

```python
result2.plot()
```
